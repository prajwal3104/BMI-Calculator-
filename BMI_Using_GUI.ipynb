```python
from tkinter import *
from tkinter import messagebox

# defining a function for reset button

def reset_entry():
    #user_tf.delete(0,'end')
    #age_lb.delete(0,'end')
    height_tf.delete(0,'end')
    weight_tf.delete(0,'end')
    milk_tf.delete(0,'end')
    vege_tf.delete(0,'end')
    meat_tf.delete(0,'end')
    lent_tf.delete(0,'end')
    egg_tf.delete(0,'end')
    rice_tf.delete(0,'end')
    
# defining a function to calculate bmi

def calculate_bmi():
    kg = int(weight_tf.get())
    m = int(height_tf.get())
    bmi = kg/(m**2)*703
    bmi = round(bmi, 2)
    bmi_index(bmi)
    
# defining a if - elif to bmi calculator

def bmi_index(bmi):
    
    if bmi <= 16:
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR', f'BMI = {bmi} is Severely Underweight')
    elif (bmi >= 16) and (bmi < 18.5):
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR', f'BMI = {bmi} is Underweight')
    elif (bmi >= 18.5) and (bmi < 25):
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR', f'BMI = {bmi} is Healthy')
    elif (bmi >=25) and (bmi < 30):
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR', f'BMI = {bmi} is Overweight') 
    elif (bmi >= 30):
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR', f'BMI = {bmi} is Obesity') 
    else:
        messagebox.showerror('bmi-CHILD NUTRITION CALCULATOR', 'something went wrong!')  

# defining a function to calculate intake of calories

def total_calorie_consumption():
    mil = int(lent_tf.get())
    veg = int(vege_tf.get())
    met = int(meat_tf.get())
    let = int(lent_tf.get())
    egg = int(egg_tf.get())
    ric = int(rice_tf.get())
    cal = int(mil*(100/100) + veg*(85/100) + met*(143/100) + let*(113/100) + egg*(155/100) + ric*(130/100))
    cal_index(cal)

# defining a if - elif to calories intake

def cal_index(cal):
    if cal >800 or cal>1400 or cal>1800 :
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR',f"The daily calorie consumption is {cal} and he (she) is not undernourished.")
    else:
        messagebox.showinfo('bmi-CHILD NUTRITION CALCULATOR',f"The daily calorie consumption of is {cal} and he (she) is  undernourished.")

# defining the visual values for gui

ws = Tk()
ws.title('CHILD NUTRITION CALCULATOR')
ws.geometry('800x600')
ws.config(bg='#8EE4AF')

var = IntVar()

# inner frame values of gui

frame = Frame(
    ws,
    padx=100, 
    pady=100,
    bg = "#EDF5E1"
)
frame.pack(expand=True)

# User Tab

user_lb = Label(
    frame,
    text='User Name',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

user_lb.grid(row=1, column=1)

user_lb = Entry(
    frame,
    bg = "#D1E8E2"
    
)
user_lb.grid(row=2, column=2, pady=5)

# Age tab

age_lb = Label(
    frame,
    text="Enter Age (Yr)",
    bg = "#66FCF1",
    font='Nunito 15 bold'
)
age_lb.grid(row=2, column=1)

age_lb = Entry(
    frame,
    bg = "#D1E8E2"
)
age_lb.grid(row=1, column=2, pady=5)

# Hight tab

height_lb = Label(
    frame,
    text="Enter Height (inc)",
    bg = "#66FCF1",
    font='Nunito 15 bold'
)
height_lb.grid(row=3, column=1)

height_tf = Entry(
    frame,
    bg = "#D1E8E2"
)

height_tf.grid(row=3, column=2, pady=5)

# Weight tab

weight_tf = Entry(
    frame,
    bg = "#D1E8E2"
)

weight_lb = Label(
    frame,
    text="Enter Weight (kg)",
    bg = "#66FCF1",
    font='Nunito 15 bold'

)
weight_lb.grid(row=4, column=1)
weight_tf.grid(row=4, column=2, pady=5)

# Button framings

frame3 = Frame(
    frame
)
frame3.grid(row=12, columnspan=3, pady=10)

# Milk tab

milk_tf = Label(
    frame,
    text='Intake of Milk (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

milk_tf.grid(row=5, column=1)

milk_tf = Entry(
    frame, 
    bg = "#D1E8E2"
)
milk_tf.grid(row=5, column=2, pady=5)

# Vegetable tab

vege_tf = Label(
    frame,
    text='Intake of Vegetable (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

vege_tf.grid(row=6, column=1)

vege_tf = Entry(
    frame,
    bg = "#D1E8E2"
)
vege_tf.grid(row=6, column=2, pady=5)

# Meat tab

meat_tf = Label(
    frame,
    text='Intake of Meat (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

meat_tf.grid(row=7, column=1)

meat_tf = Entry(
    frame, 
    bg = "#D1E8E2"
)
meat_tf.grid(row=7, column=2, pady=5)

# Bread tab

lent_tf = Label(
    frame,
    text='Intake of Bread (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

lent_tf.grid(row=8, column=1)

lent_tf = Entry(
    frame, 
    bg = "#D1E8E2"
)
lent_tf.grid(row=8, column=2, pady=5)

# Egg tab

egg_tf = Label(
    frame,
    text='Intake of Egg (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

egg_tf.grid(row=9, column=1)

egg_tf = Entry(
    frame,
    bg = "#D1E8E2"
)
egg_tf.grid(row=9, column=2, pady=5)

# Rice tab

rice_tf = Label(
    frame,
    text='Intake of Rice (cal)',
    bg = "#66FCF1",
    font='Nunito 15 bold'
)

rice_tf.grid(row=10, column=1)

rice_tf = Entry(
    frame, 
    bg = "#D1E8E2"
)
rice_tf.grid(row=10, column=2, pady=5)

# BMI calculate button

cal_btn = Button(
    frame3,
    text='Calculate BMI',
    font='Nunito 15 bold',
    command= calculate_bmi 

)
cal_btn.pack(side=LEFT)

# Reset button

reset_btn = Button(
    frame3,
    text='Reset',
    font='Nunito 15 bold',
    command=reset_entry
)
reset_btn.pack(side=LEFT)

# Calories Intake Button

exit_btn = Button(
    frame3,
    text= "Calorie's Intake",
    font='Nunito 15 bold',
    command=total_calorie_consumption
)
exit_btn.pack(side=RIGHT)

ws.mainloop()
```
